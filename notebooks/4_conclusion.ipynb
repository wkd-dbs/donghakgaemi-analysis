## 2. 핵심 발견 요약

### 발견 1: 개미 포트폴리오는 코스피에 졌다
- 5년 누적: 개미 +2.8% vs 코스피 +42.5%
- 1,000만원 투자 시 최종 차이: **397만원 손해**
- 연평균 초과수익률: **-3.2%p**

### 발견 2: 비대칭적 리스크가 핵심 원인
- 상승장 평균 초과수익: **+6.5%p** (3년)
- 하락장 평균 초과수익: **-17.8%p** (2년)
- 상승장에서 번 것을 하락장에서 3배로 토해냄

### 발견 3: 순매수 집중도가 매우 높음
- 상위 5종목이 전체 순매수의 57~81% 차지
- 삼성전자 1종목이 30~57% 차지 (2021년)
- 대형 성장주 편향 → 하락장에서 시장보다 더 큰 낙폭

### 발견 4: 고점 추격 매수 패턴
- 전년도에 많이 오른 종목을 다음 해에 매수
- 평균 회귀(mean reversion)에 취약
- 특히 2024년: 2023년 순매수 상위 종목이 -36.2% (코스피 -10.1%)

## 3. 분석의 한계

### 데이터 한계
- **순매수 금액 기준**: 순매수 수량이 아닌 금액 기준이라, 고가 종목에 편향될 수 있음
- **연간 단위 집계**: 월별/분기별 순매수 변화를 반영하지 못함
- **KOSPI만 분석**: KOSDAQ 종목은 포함되지 않음
- **배당 미반영**: 종가 기준 수익률만 계산, 배당 수익 미포함

### 방법론 한계
- **전년도 기준 포트폴리오**: 연중 순매수 변화를 반영하지 않는 단순 전략
- **동일 비중 가정**: 실제 개인 투자자는 순매수 금액에 비례하여 투자하지 않음
- **거래 비용 미반영**: 매매 수수료, 세금 미포함
- **5년 기간**: 통계적 유의성을 확보하기에 표본 기간이 짧음

### 해석상 주의
- 본 분석은 **"개인 투자자 순매수 상위 종목을 따라 사는 전략"**의 성과를 검증한 것이지, 개인 투자자 전체의 실제 수익률과는 다름
- 개인 투자자 중에서도 시장을 이긴 투자자가 있을 수 있으나, "군중을 따라가는 전략"은 시장을 이기지 못함을 보여줌

## 4. 향후 개선 방향

- **월별 리밸런싱**: 연간이 아닌 월별 순매수 데이터로 포트폴리오 재구성
- **KOSDAQ 포함**: KOSDAQ 개인 순매수 종목까지 분석 확대
- **역발상 전략**: 개미 순매도 상위 종목(=기관/외국인 순매수)의 성과와 비교
- **배당 반영**: 총수익률(Total Return) 기준으로 재계산
- **위험 조정 수익률**: 샤프 비율(Sharpe Ratio)로 리스크 대비 성과 비교
- **기간 확장**: 2015~2019년 데이터 추가로 표본 기간 확대
